```python
# Diketahui KOH dititrasi dengan HNO3
# KOH adalah basa kuat menjadi titrat dan HNO3 adalah asam kuat menjadi titran

konsentrasi_HNO3 = 0.05 # M
volume_HNO3_list = [0, 10, 20, 25, 29.5, 29.9, 29.95, 30, 30.05, 30.1, 39, 40] # mL
volume_KOH = 30.0 # mL
konsentrasi_KOH = 0.05 # M
```

```python
# Input Data pH dan Perhitungan
pd.read_csv ('/KAKKVI.csv')
```

```python
# Menghitung jumlah mol KOH awal
mol_KOH_awal = konsentrasi_KOH * volume_KOH / 1000 # konversi mL ke L
```

```python
print("Volume HNO3 (mL) | pH")
print("------------------|-------")

# Fungsi untuk menghitung pH
def hitung_pH(volume_HNO3):
    mol_HNO3 = konsentrasi_HNO3 * volume_HNO3 / 1000 # konversi mL ke L

    if mol_HNO3 < mol_KOH_awal:
        mol_KOH_berlebih = mol_KOH_awal - mol_HNO3
        konsentrasi_OH = mol_KOH_berlebih / ((volume_HNO3 + volume_KOH) / 1000) # konversi total volume ke L
        pOH = -math.log10(konsentrasi_OH)
        pH = 14 - pOH
    elif mol_HNO3 == mol_KOH_awal:
        pH = 7.0  # titik ekivalen, netral
    else:
        mol_HNO3_berlebih = mol_HNO3 - mol_KOH_awal
        konsentrasi_H = mol_HNO3_berlebih / ((volume_HNO3 + volume_KOH) / 1000) # konversi total volume ke L
        pH = -math.log10(konsentrasi_H)
    return pH

# Menghitung pH untuk setiap volume HNO3 yang ditambahkan
pH_list = []
for volume_HNO3 in volume_HNO3_list:
    pH = hitung_pH(volume_HNO3)
    pH_list.append(pH)
    print(f"{volume_HNO3:<18} | {pH:.2f}")
```

```python
# Membuat grafik pH vs volume HNO3
plt.figure(figsize=(10, 6))
plt.plot(volume_HNO3_list, pH_list, marker='o', color='blue', label='pH Titrasi HNO3 dengan KOH')
plt.axhline(y=7.0, color='gray', linestyle='--', linewidth=1.5, label='Titik Ekuivalen (pH 7)')
plt.axvline(x=30.0, color='red', linestyle='--', linewidth=1.5, label='Volume Ekuivalen')
plt.title('Kurva Titrasi HNO3 vs KOH')
plt.xlabel('Volume HNO3 (mL)')
plt.ylabel('pH')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Menghitung pH untuk setiap volume HNO3 yang ditambahkan dan menentukan warna indikator BTB
warna_list_btb = []
for volume in volume_HNO3_list:
    pH = hitung_pH(volume)
    pH_list_btb.append(pH)
    if pH < 6.0:
        warna_list_btb.append('yellow')
    elif 6.0 <= pH < 7.6:
        warna_list_btb.append('green')
    else:
        warna_list_btb.append('blue')
```

```python
# Membuat grafik pH vs volume HNO3 dengan indikator Bromtimol Biru
plt.figure(figsize=(10, 6))
for i in range(len(volume_HNO3_list) - 1):
    plt.plot(volume_HNO3_list[i:i+2], pH_list_btb[i:i+2], marker='o', color=warna_list_btb[i])

plt.axhline(y=7.0, color='gray', linestyle='--', linewidth=1.5, label='Titik Ekuivalen (pH 7)')
plt.axvline(x=30.0, color='red', linestyle='--', linewidth=1.5, label='Volume Ekuivalen')
plt.title('Kurva Titrasi HNO3 vs KOH dengan Indikator Bromtimol Biru')
plt.xlabel('Volume HNO3 (mL)')
plt.ylabel('pH')
plt.legend(loc='upper left')
plt.grid(True)
plt.show()
```
