# **Visualizations to Explain Data**

Bagian ini berisi visualisasi data yang dapat membantu dalam mengidentifikasi dan mendeteksi masalah, serta mempermudah interpretasi data bagi pembaca



```python
# Rancang Histogram untuk Setiap Kolom dari Data
data.hist(figsize=(14, 12), color=sns.color_palette('hls')[0])
plt.suptitle('Histograms Setiap Kolom')
plt.show()
```

```python
# Lakukan Analisis Bivariat
def student_info(x):
   return data.groupby([x])['StudentID'].count()
student_info('Gender')
#0 => male
#1 => female

# Rancang Bar Chart
plt.figure(figsize= (3,5))
plt.bar(['Female','Male'],data["Gender"].value_counts().values,color= ['palevioletred','royalblue'])
plt.title('Distribusi Gender')
plt.ylabel('Jumlah')
plt.show()
```

```python
# Visualisasi Berbagai Kelompok Etnis dalam Dataset
ethnicity_labels = ['Caucasian', 'African American', 'Asian', 'Others']

plt.figure(figsize=(12, 6))
ax = sns.countplot(x='Ethnicity', data=data, palette='Set2')
plt.title('Distribusi Suku Bangsa')
plt.xlabel('Suku Bangsa')
plt.ylabel('Jumlah')
plt.xticks(ticks=range(len(ethnicity_labels)), labels=ethnicity_labels, rotation=45)

for p in ax.patches:
    height = p.get_height()
    ax.annotate(f'{int(height)}',
                (p.get_x() + p.get_width() / 2., height),
                ha='center', va='center',
                xytext=(0, 5),
                textcoords='offset points')

plt.show()
```

> **Activity and GPA Correlation**

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='Sports', y='GPA', hue='Music', data=data, palette={0: "plum", 1: "lightpink"})
plt.title('Nilai GPA Berdasarkan Partisipasi Olahraga dan Musik')
plt.xlabel('Sports Participation (0: No, 1: Yes)')
plt.ylabel('GPA')
plt.show()
```

```python
# Visualisasi Pengaruh Partisipasi Olahraga Terhadap Nilai GPA
fig = px.box(data, x='Sports', y='GPA', title='GPA dan Partisipasi Olahraga',
             labels={'Sports': 'Sports Participation (0: No, 1: Yes)', 'GPA': 'GPA'},
             color='Sports')
fig.show()
```

```python
# Rancang Pair Plot
plt.figure(figsize=(12, 6))
sns.pairplot(data, vars=['GPA', 'Sports', 'Music', 'Volunteering'],
                 kind='scatter',  # Pastikan menggunakan scatter plot
                 plot_kws={'color': 'mediumblue', 'edgecolor': 'k'})  # Atur warna dan edgecolor
plt.suptitle('Hubungan antara Kegiatan Ekstrakurikuler dan Nilai GPA', y=1)
plt.show()

# Tampilkan Korelasi antara Kegiatan Ekstrakurikuler dan Nilai GPA
sports_gpa_corr = data['Sports'].corr(data['GPA'])
music_gpa_corr = data['Music'].corr(data['GPA'])
volunteering_gpa_corr = data['Volunteering'].corr(data['GPA'])
print(f'Sports-GPA correlation: {sports_gpa_corr:.2f}')
print(f'Music-GPA correlation: {music_gpa_corr:.2f}')
print(f'Volunteering-GPA correlation: {volunteering_gpa_corr:.2f}')
```

> **Study Habits**

```python
# Hubungan antara Waktu Belajar dan Nilai GPA Menggunakan Trendline
fig = px.scatter(data, x='StudyTimeWeekly', y='GPA', trendline='lowess',
                 labels={'StudyTimeWeekly': 'Study Time Weekly (hours)', 'GPA': 'GPA'},
                 title='Waktu Belajar Mingguan vs Nilai GPA Menggunakan Trendline',
                 color_discrete_sequence=["turquoise"])

# Kustomisasi Plot
fig.update_layout(
    xaxis_title='Study Time Weekly (hours)',
    yaxis_title='GPA',
    legend_title='Legend'
)

fig.show()
```

> **Parental Influence**

```python
# Visualisasi Rata-rata Dukungan Orang Tua di Seluruh Jenjang Pendidikan
plt.figure(figsize=(12, 6))
avg_support = data.groupby('ParentalEducation')['ParentalSupport'].mean().reset_index()

plt.plot(avg_support['ParentalEducation'], avg_support['ParentalSupport'], marker='o', linestyle='-', color='black')
plt.title('Rata-rata Dukungan Orang Tua Berdasarkan Tingkat Pendidikan Orang Tua')
plt.xlabel('Parental Education Level')
plt.ylabel('Average Parental Support')
plt.grid(True)
plt.show()
```

```python
# Visualisasi dengan Absensi
plt.figure(figsize=(12, 6))
sns.scatterplot(x='ParentalSupport', y='Absences', data=data)
plt.title('Hubungan antara Dukungan Orang Tua dan Ketidakhadiran')
plt.xlabel('Parental Support')
plt.ylabel('Absences')
plt.show()
support_absences_corr = data['ParentalSupport'].corr(data['Absences'])
print(f'Correlation coefficient between Parental Support and Absences: {support_absences_corr:.2f}')
```

> **Correlation Matrix**

Berisi bagian visualisasi matriks korelasi data dalam bentuk heatmap guna memahami hubungan antarvariabel dalam dataset

```python
# Pengaruh Dukungan Orang Tua dan Kegiatan Ekstrakurikuler Terhadap Nilai GPA
plt.figure(figsize=(12, 6))
heatmap_data = pd.crosstab(data['ParentalSupport'], data['Extracurricular'], values=data['GPA'], aggfunc='mean')
sns.heatmap(heatmap_data, cmap='Spectral', annot=True)
plt.title('Dukungan Orang Tua dan Kegiatan Ekstrakurikuler vs GPA')
plt.xlabel('Extracurricular (0: No, 1: Yes)')
plt.ylabel('Parental Support Level')
plt.show()
```

```python
# Heat Map Setiap Kolom
correlation_matrix = data.corr()
plt.figure(figsize=(10, 10))
sns.heatmap(correlation_matrix,annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Matriks Korelasi')
plt.show()
```
